```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a Panda's DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first five rows of the dataset
credit_card_data.head()
```

```python
#last five rows of the dataset
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the numbers of missing values in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fraudelent transaction
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced.

The label 0 represents legit transacrion
The label 1 represents fraudelent transaction

```python
#separating the data for anaylsis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaactions
credit_card_data.groupby('Class').mean()
```

Under Sampling

Build a similar dataset containing equal amounts of legit and fraud transactions

number of fraud transactions : 492

So adding 492 random legit transactions, to have good dataset as we have even distribution.

```python
legit_sample = legit.sample(n=492)
```

Concatenating two data frames

```python
new_dataset= pd.concat([legit_sample, fraud],axis=0)
#axis = 0 --> means rows
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets


```python
X= new_dataset.drop(columns='Class',axis = 1)
#axis = 1 means rows
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitiing the data into Testing Data and Training Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, stratify=Y , random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

Training the Logistic Regression Model with Training Data

```python
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training Data : ' , training_data_accuracy )
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy Score on test data : ', test_data_accuracy)
```
